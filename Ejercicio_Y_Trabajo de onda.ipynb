Ondas generadas por una piedra 

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("Ingrese la posición de x: "))
y1 = float(input("Ingrese la posición de y: "))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        outfile.write("%0.1f " % (zeta0*sin(k*r) ) )
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("Distacian en x en cm")
ylabel("Distacian en y en cm")
imshow(data, origin="lower", extent=[0, 100, 0, 100])
show()
```

Ondas generadas por dos piedras distanciadas a 20 cm 

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("Ingrese la posición de x: ")) # cm
y1 = float(input("Ingrese la posición de y: ")) # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r2 = sqrt( ((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        t=zeta0*sin(k*r)
        t2=zeta0*sin(k*r2)
        outfile.write("%0.1f " % (t2) )
        outfile.write("%0.1f " % (t) )
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("Distacian en x en cm")
ylabel("Distacian en y en cm")
imshow(data, origin="lower", extent=[0, 100, 0, 100])
show()
```
